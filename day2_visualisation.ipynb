```
def group_and_barplot(feat_groupby, feat_agg='price_value', feat_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
  return (
    df
      .groupby(feat_groupby)[feat_agg]
      .agg(feat_funcs)
      .sort_values(by=feat_sort, ascending = False)
      .head(top)
).plot(kind='bar', figsize=(15,5), subplots=subplots)
```

```
group_and_barplot('param_marka-pojazdu');
```

```
group_and_barplot('param_kraj-pochodzenia');
```

```
group_and_barplot('param_rok-produkcji',feat_sort='size');
```

```
group_and_barplot('param_kolor',feat_sort='mean');
```
